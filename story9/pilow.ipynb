# **creations filtre**

```python
def fltr_rouge(img):
    imgr=np.copy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            r,v,b=img[i,j]
            imgr[i,j]= (r,0,0)
    return imgr

def fltr_bleu(img):
    imgb=np.copy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            r,v,b=img[i,j]
            imgb[i,j]= (0,0,b)
    return imgb

def fltr_vert(img):
    imgv=np.copy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            r,v,b=img[i,j]
            imgv[i,j]= (0,v,0)
    return imgv
```

```python
imgpilr=Image.fromarray(fltr_rouge(img))
imgpilr.save(r"photos\foretrouge.jpg")
```

```python
imgpilb=Image.fromarray(fltr_bleu(img))
imgpilb.save(r"photos\foretbleu.jpg")
```

```python
imgpilv=Image.fromarray(fltr_vert(img))
imgpilv.save(r"photos\foretvert.jpg")
```

```python
def filtre(img_orig):
    im = np.copy(img_orig) # On fait une copie de l'original
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            r, v, b = im[i, j]
            im[i, j] = (99-r, 99-v,99-b)
    return im
```

```python
imgpilf=Image.fromarray(filtre(img))
imgpilf.save(r"photos\foretf.jpg")
```

```python
from math import exp,pi 
 
def gaussKernel(r,sigma):
    kernel = [[0] * (2*r+1) for _ in range(2*r + 1)] #creation d'une matrice carée vide de coté impaire
    if sigma > 0:
        gaussianKernelFactor, e = 0, 0 #init des valeur du total et de la valeur de chaque case
        for ky in range(-r,r):
            for kx in range(-r,r):
                e = exp(-(kx**2+ky**2)/(2*pi*sigma**2)) #calcule de la valeur avec la formul de gausse
                gaussianKernelFactor += e # somme de toute les valeur au fur et a mesur
                kernel[kx+r][ky+r] = e #inscription dan la matrice
        
        for ky in range(-r,r):
            for kx in range(-r,r):
                kernel[kx+r][ky+r] = kernel[kx+r][ky+r] / gaussianKernelFactor #normalisation des valeur en divisant pas le total pour avoir une somme egal a un
                
    return kernel
                
def convolution(img,M):
    px = int( (len(M)-1) / 2 ) # mesure de la perte sur les bords taille du bord
    im = img.copy()
    imax = img.shape[1] - px #taille de limage apres passage du kernel ou range du kernel
    for i in range(px,imax):
        for j in range(px,img.shape[0]-px):# on aurais pus le metre dan une variable jmax=img.shape[0]-px
            somme = 0.0
            for k in range(-px,px+1):
                for l in range(-px,px+1):
                    somme += img[j+l][i+k] * M[l+px][k+px]
            im[j][i] = somme
    return im
imgpil = Image.open(r"photos\foret.jpg")  
img = np.array(imgpil) 
H = gaussKernel(3,10)
imgpil = Image.fromarray(convolution(img,H))
imgpil.save("resultat_flou2.jpg")
```
