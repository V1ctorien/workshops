```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("https://www.stat4decision.com/wine_data.csv", index_col = 0)
data["type"].value_counts()
```

```python
data.info()
```

```python
y = data["type"] # target
x = data.drop("type", axis = 1) # features
x_train, x_test, y_train, y_test  = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
modele_rf = RandomForestClassifier(
     n_estimators=100, # nombre de modèles dans la forêt
     criterion='gini', # critère de séparation
     max_depth=None, # profondeur maximale des arbres
     min_samples_split=2, # nombre minimum d'échantillons pour diviser un nœud
     min_samples_leaf=1, # nombre minimum d'échantillons dans une feuille
     min_weight_fraction_leaf=0.0, # fraction minimale du poids total des échantillons requis dans une feuille
      # max feature nombre de caractéristiques à considérer pour la meilleure séparation
     max_leaf_nodes=None, # nombre maximum de nœuds feuilles
     min_impurity_decrease=0.0, # seuil de diminution d'impureté pour diviser un nœud
     bootstrap=True, # utilisation de l'échantillonnage bootstrap
     oob_score=False, # utilisation des échantillons hors sac pour estimer la performance
     n_jobs=None, # nombre de jobs à utiliser pour l'entraînement
     random_state=None, # graine aléatoire
     verbose=0, # niveau de verbosité
     warm_start=False, # réutilisation des solutions précédentes pour ajuster et ajouter plus d'arbres
     class_weight=None, # poids des classes
     ccp_alpha=0.0, # paramètre de complexité pour la post-élagage
     max_samples=None,) # nombre d'échantillons à tirer pour chaque arbre lors de l'utilisation de bootstrap
```

```python
modele_rf.fit(x_train, y_train)

pd.DataFrame(modele_rf.feature_importances_,
              index = x_train.columns, 
              columns = ["importance"]).sort_values(
     "importance", 
     ascending = False)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(f"Le pourcentage de bien classés est de : {accuracy_score(y_test, modele_rf.predict(x_test))*100} %")
```

```python
pd.DataFrame(confusion_matrix(y_test, modele_rf.predict(x_test)), # les params de confusion_matrix sont (valeurs vraies, valeurs prédites)
             index = ["blanc_données", "rouge_données"], # index c'est les vraies valeurs
             columns = ["blanc_predit", "rouge_predit"]) # colonnes c'est les valeurs prédites
```
